## Cleaning and Preprocessing

### Creating separated columns for city and state


```python
# To better understand the data, we will split the column 'city_state' into 
# two separate columns: 'city' and 'state'. With this, we can analyze the leads separately by state and city.

# Creating a new column 'city' and 'state' from 'city_state'
for _,row in df_raw.iterrows():
    str_cs = row['city_state']
    cs_ = str_cs.split('_')
    df_raw.at[_,"city_state"] = str_cs  
    df_raw.at[_,"city"] = cs_[1].strip()
    df_raw.at[_,"state"] = cs_[0].strip()
# Drop the original 'city_state' column
df_raw.drop(columns=['city_state'], inplace=True)

# Function to fix encoding issues in text columns
def fix_encoding(text):
    if isinstance(text, str):
        try:
            return (
                text
                .encode('latin1', errors='replace')
                .decode('utf-8', errors='replace')
            )
        except:
            return text
    return text



df_raw['city'] = df_raw['city'].apply(fix_encoding)
```

### Transforming flag columns to int

```python
# This block will check if there are any columns that start with 'flg_' and convert them to integers. 
# As we saw in previous analysis these flg columns are categorical variables that indicate the 
# presence of a feature, so we will convert them to integers (1 for 'Y' and 0 for 'N').
for col in df_raw.columns:
    if col.startswith("flg_"):
        df_raw[col] = df_raw[col].fillna(0)
        df_raw[col] = df_raw[col].replace({"S": "Y"}).replace({"Y": 1, "N":0}).astype(int)
```

### Transforming fuel type columns in multiple columns

```python
df_raw.fuel_type.unique()
```

```python
# These type of glimpse you will acquire with experience, but it is always good to check the unique values of a column
# to understand the data better. In this case, we can see that the fuel_type column
# has multiple fuel types in a single string, separated by spaces and commas.
# This is not ideal for analysis, as we want to have each fuel type in a separate column to better 
# understand how the fuel type can influence on the leads

# The goal here is to transform the fuel_type column into multiple columns,
# one for each fuel type as a flag column

# First, we will create a list of unique fuel types and find the unique values in the 'fuel_type' column

df_processed = df_raw.copy()
fuel_types = []
for item in df_processed.fuel_type.unique():
    if isinstance(item, str): # check if item is a string
        # Split the string by spaces and commas, and strip any whitespace
        str_lst = item.split(' ')
        for i in range(len(str_lst)):
            str_lst[i] = str_lst[i].strip().replace(',', '') # remove commas
            if str_lst[i] != 'e': # check if the item is not 'e'
                if str_lst[i] not in fuel_types: # check if the item is not already in the list
                    fuel_types.append(str_lst[i])

print("These are all fuel types:", fuel_types)

# Second, now we know the unique fuel types, we can create a new column for each fuel type
# But first, we need to fix the two type 'natural' and 'gas' to be the same 'natural gas'
fuel_types.remove('natural')
fuel_types.remove('gas')
fuel_types.append('gas natural')

df_processed['fuel_type'] = df_processed['fuel_type'].fillna('')  # Fill NaN values with empty string to avoid errors

# Third, we need to create a new column for each fuel type
for fuel in fuel_types:
    df_processed[f"flg_{fuel.replace(" ", "_")}"] = df_processed['fuel_type'].apply(lambda x: 1 if fuel in x else 0 if x == '' else 0)

# Now we can drop the original fuel_type column
df_processed.drop(columns=['fuel_type'], inplace=True)
```

```python
def summarize_numerical_features(df, cols, target_col=None):
    num_cols = [col for col in cols if col != target_col]

    summary = df[num_cols].describe().T
    summary['skew'] = df[num_cols].skew()
    summary['missing_pct'] = df[num_cols].isnull().mean() * 100
    
    if target_col:
        summary['corr_with_target'] = df[num_cols].corrwith(df[target_col])
        return summary.sort_values(by='corr_with_target', ascending=False)
    else:
        return summary

numerical_summary = summarize_numerical_features(df_processed, cols=num_cols, target_col='leads')
display(numerical_summary)
```

```python
def remove_outliers_iqr(df, columns, k=1.5):
    df_clean = df.copy()
    for col in columns:
        q1 = df[col].quantile(0.01)
        q3 = df[col].quantile(0.995)
        iqr = q3 - q1
        lower = q1 - k * iqr
        upper = q3 + k * iqr
        print(f"Removing outliers for {col}: IQR = {iqr}, lower bound = {lower}, upper bound = {upper}")
        df_clean = df_clean[(df_clean[col] >= lower) & (df_clean[col] <= upper)]
    return df_clean

df_clean = remove_outliers_iqr(df_processed, num_cols).reset_index(drop=True)

fig, axes = plt.subplots(1, n_cols, figsize=(n_cols*4, 5), constrained_layout=True)

for i, col in enumerate(num_cols):
    ax = axes[i] if n_cols > 1 else axes
    sns.boxplot(data=df_clean, x=col, ax=ax)
    ax.set_title(f'Boxplot of {col}')
    ax.set_xlabel(col)

plt.suptitle('Boxplots of Numerical Features', fontsize=16, y=1.02)
plt.show()
```

## Categorical Features Calssification

- This type of analysis is essential to understand what kind of categorical variables we are dealing with and how to properly use them in model development.

- Categorical variables often contain a large number of unique classes — for example, a variable like city can have hundreds of distinct locations. Feeding such high-granularity variables directly into the model may not be effective.

- High cardinality or high granularity can bias the model. In such cases, the model might learn to always give the same prediction for a specific class (e.g., always output a high lead count for a particular city), regardless of other features. This happens because the model assigns disproportionate importance to the most frequent or highly specific categories, reducing generalization and increasing the risk of overfitting.

```python
# This is an important step of the EDA process, as it helps to understand the nature of 
# categorical features in the dataset. We will classify categorical features into:
# - Binary: Features with exactly two unique values (e.g., Yes/No, True/False).
# - Nominal (Low Cardinality): Features with a small number of unique values (e.g., colors, types).
# - Nominal (Medium Cardinality): Features with a moderate number of unique values (e.g., cities, brands).
# - Nominal (High Cardinality): Features with a large number of unique values (e.g., user IDs, product codes).  

cat_cols.append("priority")


def classify_categorical_features(
    df,
    cols,
    threshold_low=15,
    threshold_high=50,
    id_keywords=["id", "cd_", "code"],
):
    result = []

    for col in cols:
        if (
            df[col].dtype == "object" or df[col].nunique() < 1000
        ):  
            nunique = df[col].nunique()
            col_lower = col.lower()

            if nunique == 2:
                category_type = "binary"
            elif any(key in col_lower for key in id_keywords):
                category_type = "identity/high cardinality"
            elif nunique <= threshold_low:
                category_type = "nominal (low cardinality)"
            elif nunique <= threshold_high:
                category_type = "nominal (medium cardinality)"
            else:
                category_type = "nominal (high cardinality)"

            result.append(
                {
                    "feature": col,
                    "dtype": str(df[col].dtype),
                    "n_unique": nunique,
                    "category_type": category_type,
                }
            )

    return pd.DataFrame(result).sort_values(by="category_type")


summary = classify_categorical_features(df_clean, cols=cat_cols)
summary
```

## Bivariate Analysis 

```python
df_biv = df_clean.copy()
dfs_bivariate = {}

for col in cat_cols:
    grouped = (
        df_biv.groupby(col)['leads']
        .agg(
            count='count',
            mean_leads='mean',
            sum_leads='sum'
        )
        .sort_values(by='mean_leads', ascending=False)
        .reset_index()
    )
    grouped['percentage'] = (grouped['count'] / grouped['count'].sum()) * 100
    dfs_bivariate[col] = grouped
```

```python
for keys in dfs_bivariate.keys():
    if keys.startswith('flg_') or keys.startswith('cd_'):
        continue
    else:
        print(f"Top 10 categories for {keys}:")
        display(dfs_bivariate[keys].sort_values(by='sum_leads', ascending=False).head(10))
        print("\n")
```

```python

def plot_top_categories(col_name, metric='sum_leads', top_n=10):
    df_plot = dfs_bivariate[col_name].sort_values(by=metric, ascending=False).head(top_n)
    sns.barplot(x=metric, y=col_name, data=df_plot)
    plt.title(f"{col_name} - Top {top_n} by {metric}")
    plt.show()

plot_top_categories('city')
```

```python
def binned_target_analysis(df, feature, target, bins=10):
    df_cut = pd.cut(df[feature], bins=bins)
    return df.groupby(df_cut)[target].agg(['count']).sort_index()

higher_bin = {}
for col in num_cols:
    if col != 'leads':
        binned_analysis = binned_target_analysis(df_clean, col, 'leads', bins=10)
        higher_bin[col] = {"max": binned_analysis['count'].max(),
                           "proportion_max": binned_analysis['count'].max() / binned_analysis['count'].sum(),
                           "min": binned_analysis['count'].min(),
                           "bin_counts": binned_analysis['count'].idxmax()}
higher_bin
```

```python
import xgboost as xgb
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder

df = df_clean.copy()

df[num_cols] = df[num_cols].fillna(0)

cols_ = [
    "views",
    "phone_clicks",
    "vl_advertise",
    "n_photos",
    "km_vehicle",
    "vl_market",
    "n_doors"
]  

X = df[cols_]
y = df["flg_leads"] 

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=42)


model = xgb.XGBClassifier(
    n_estimators=100,
    max_depth=6,
    learning_rate=0.01,
    random_state=42,
    tree_method="auto",  # fast on large datasets
)

model.fit(X_train, y_train)
```
